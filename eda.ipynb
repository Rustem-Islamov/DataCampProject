```python
data_2017 = pd.read_csv(config.path_2017)
data_2017 = pd.read_csv(os.path.join('data','PR17_BVot_T1_FE.txt'),sep=';',encoding='latin1')
```

```python
data_2017_first_half = data_2017.index.to_frame().reset_index()
data_2017_second_half = data_2017.reset_index(drop=True)
```

```python
columns_infos = ['Code du département', 'Libellé du département',
'Code de la circonscription', 'Libellé de la circonscription',
'Code de la commune', 'Libellé de la commune', 'Code du b.vote',
'Inscrits', 'Abstentions', '% Abs/Ins', 'Votants', '% Vot/Ins',
'Blancs', '% Blancs/Ins', '% Blancs/Vot', 'Nuls', '% Nuls/Ins',
'% Nuls/Vot', 'Exprimés', '% Exp/Ins', '% Exp/Vot', ]
columns_to_be_repeated = ['N°Panneau', 'Sexe','Nom', 'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp']
data_2017_second_half.columns = [i for i in range(70,70+data_2017_second_half.shape[1])]
data_final_2017 = pd.concat([data_2017_first_half, data_2017_second_half],axis=1)
```

```python
number_repeats = (data_final_2017.shape[1] - len(columns_infos))/len(columns_to_be_repeated)
```

```python
data_final_2017.columns = columns_infos + [f'{element}_{i}'for i in range(1,int(number_repeats)+1) for element in columns_to_be_repeated ]
```

```python
df_2017  = pd.DataFrame(columns=columns_infos+columns_to_be_repeated)
n=0
for i in range(int(number_repeats)):
    n = len(columns_infos)+(i)*len(columns_to_be_repeated)
    indexes = [i for i in range(len(columns_infos))] + [i for i in range(n,n+len(columns_to_be_repeated))]
    data = data_final_2017.iloc[:,indexes]
    data.columns = columns_infos+columns_to_be_repeated
    df_2017 = pd.concat([df_2017, data])
```
