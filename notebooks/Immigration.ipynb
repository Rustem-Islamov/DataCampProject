```python
data_immigration = pd.read_csv('BTT_TD_NAT1_2018.CSV',sep=';')
```

```python
data_immigration[data_immigration['INATC']==2].groupby(['LIBGEO'])["NB"].sum().sort_values(ascending=False)
```

```python
paris_arrons =[element for element in data_immigration["LIBGEO"].unique() if "Paris" in element][:20]
```

```python
data_immigration[(data_immigration['INATC']==2)&(data_immigration["LIBGEO"].isin(paris_arrons))].groupby(['LIBGEO'])["NB"].sum().sort_values(ascending=False)/data_immigration[(data_immigration["LIBGEO"].isin(paris_arrons))].groupby(['LIBGEO'])["NB"].sum().sort_values(ascending=False).values
```

```python
data_immigration[(data_immigration['INATC']==1) & (~(data_immigration["LIBGEO"].isin(paris_arrons)))]['NB'].sum()
```

```python
data_immigration["Dep"]=data_immigration["CODGEO"].apply(lambda element:str(element)[:2])
```

```python
villes=pd.DataFrame(data_immigration[~(data_immigration["LIBGEO"].isin(paris_arrons))].groupby(["LIBGEO"])['NB'].sum()).reset_index()
noms_villes=list(villes[villes['NB']>1000]['LIBGEO'].unique())
```

```python
(data_immigration[(data_immigration['INATC']==2) & 
                 (~(data_immigration["LIBGEO"].isin(paris_arrons)))].groupby(["LIBGEO"])['NB'].sum()/data_immigration[~(data_immigration["LIBGEO"].isin(paris_arrons))].groupby(["LIBGEO"])['NB'].sum()).sort_values(ascending=False).iloc[:20]
```

```python
(data_immigration[(data_immigration['INATC']==2) & 
                 (data_immigration['LIBGEO'].isin(noms_villes)) 
                  &(data_immigration['Dep']!='97')].groupby(["LIBGEO"])['NB'].sum()/data_immigration[(data_immigration['LIBGEO'].isin(noms_villes))
                         &(data_immigration['Dep']!='97')].groupby(["LIBGEO"])['NB'].sum()).sort_values(ascending=False).iloc[:20]
```
